## 1. Lets DO Analysis for BIHAR & JHARKHAND

```python
air = pd.read_csv('Airtel21-23.csv', index_col='State')
```

```python
air = air[['Jan23', 'Feb23', 'Mar23', 'Apr23', 'May23', 'Jun23', 'Jul23',
       'Aug23', 'Sep23', 'Nov23', 'Dec23']]
```

```python
air.loc['Bihar'].plot(kind= 'bar')
air.loc['Bihar'].plot(kind= 'line', marker = 'o', color='green')

plt.title('Airtel users in Bihar & Jharkhand in year 2023')
plt.xlabel('Month wise')
plt.ylabel('Total telecom users')
plt.xticks(rotation =45)
plt.ticklabel_format(style='plain', axis='y')
plt.ylim(39000000)
```

The above graph captures the number of telecom users in Bihar & Jharkhand, and we see there is a steady increase throughout year 2023

```python
Airtel.loc['Bihar'].plot(kind = 'line', marker = 'o', label = 'Airtel')
Jio.loc['Bihar'].plot(kind = 'line', marker = 'o', label = 'Jio')
VI.loc['Bihar'].plot(kind = 'line', marker = 'o', label = 'VI')
BSNL.loc['Bihar'].plot(kind = 'line', marker = 'o', label = 'BSNL')

plt.xlabel('Month wise')
plt.ylabel('% distribution')
plt.title('Network providers for year 2023 in Bihar & Jharkhand')
plt.legend()
```

In the above graph here, the following can be observed:

Airtel and Jio: Both Airtel and Jio show an increasing trend in their market share throughout 2023. This could be due to various factors such as network upgrades, better customer service, attractive pricing strategies, or successful marketing campaigns that resonate with the local population.

VI and BSNL: On the other hand, both VI and BSNL show a decline in their market share. This could be due to factors such as poor network quality, higher pricing, or lack of effective marketing strategies.

Competition: The graph also shows a similar competition and market growth between Airtel and Jio in Bihar. This indicates that both companies are investing heavily in this region and are likely offering competitive services to gain market share.

Market Share: By the end of 2023, both VI and BSNL will have less than a 5% market share in Bihar, the second most populous state in the country. This suggests that these companies may need to rethink their strategies to regain their market share in this region.
