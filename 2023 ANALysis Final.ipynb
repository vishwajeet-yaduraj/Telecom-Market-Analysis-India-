## 1. Lets DO Analysis for BIHAR

```python
air = pd.read_csv('Airtel21-23.csv', index_col='State')
```

```python
air = air[['Jan23', 'Feb23', 'Mar23', 'Apr23', 'May23', 'Jun23', 'Jul23',
       'Aug23', 'Sep23', 'Nov23', 'Dec23']]
```

```python
air.loc['Bihar'].plot(kind= 'bar')
air.loc['Bihar'].plot(kind= 'line', marker = 'o', color='green')

plt.title('Airtel users in Bihar(Jharkhand) in year 2023')
plt.xlabel('Month wise')
plt.ylabel('Total telecom users')
plt.xticks(rotation =45)
plt.ticklabel_format(style='plain', axis='y')
plt.ylim(39000000)
```

The above graph captures the number of telecom users in Bihar(Jharkhand included) and we see there is a constant increase throughout year 2023

```python
Airtel.loc['Bihar'].plot(kind = 'line', marker = 'o', label = 'Airtel')
Jio.loc['Bihar'].plot(kind = 'line', marker = 'o', label = 'Jio')
VI.loc['Bihar'].plot(kind = 'line', marker = 'o', label = 'VI')
BSNL.loc['Bihar'].plot(kind = 'line', marker = 'o', label = 'BSNL')

plt.xlabel('Month wise')
plt.ylabel('% distribution')
plt.title('Network providers for year 2023 in Bihar(Jharkhand)')
plt.legend()
```

In the above graph here, it is observed that while Jio along with Airtel seem to have similar competition and market.
Both of them show increasing curve for telecom users in eastern state of Bihar.

And the other two BSNL and VI have declining curve and no. of users in Bihar(Jharkhand).

By the end of 2023, Both of them have less than 5% market share in Bihar(second most popolous state in the country)

```python
Others = 100 - (Airtel.loc['Bihar'][0] + Jio.loc['Bihar'][0] + VI.loc['Bihar'][0] + BSNL.loc['Bihar'][0])
other = 100 - (Airtel.loc['Bihar'][-1] + Jio.loc['Bihar'][-1] + VI.loc['Bihar'][-1] + BSNL.loc['Bihar'][-1])
```

```python
Bihar_start = [Airtel.loc['Bihar'][0],Jio.loc['Bihar'][0],VI.loc['Bihar'][0],BSNL.loc['Bihar'][0], Others]
Bihar_end = [Airtel.loc['Bihar'][-1],Jio.loc['Bihar'][-1],VI.loc['Bihar'][-1],BSNL.loc['Bihar'][-1], other]
```

```python
Bihar_start
```

```python
plt.figure(figsize=(10, 6))
plt.pie(Bihar_start, labels = ['Airtel', 'Jio', 'VI', 'BSNL', 'Others'], autopct='%.2f')
plt.title('Pie chart distribution of Bihar(Jharkhand) at start of 2023')
```

The above pie chart shows the distribution of user in Bihar for year 2023 in an excellent manner.

Some observations from the graph above:
1. Airtel and Jio have the highest market share.
2. Airtel seem to be ahead of Jio by start of 2023, with difference of more than 2%.
3. VI and BSNL have very little market shares 
4. And the population who are not using any telecom seem to be huge with above 45%. Such a big amount of population have no telecom subscriptions. { What can be the reason, is the number similar to other states. We have to look into possible reasons}

```python
plt.figure(figsize=(10, 6))
plt.pie(Bihar_end, labels = ['Airtel', 'Jio', 'VI', 'BSNL', 'Others'], autopct='%.2f')
plt.title('Pie chart distribution of Bihar(Jharkhand) at end of 2023')
```

The above pie chart is for the end of 2023

Some key points here:
1. Jio seem to increase in much faster rate than Airtel, as the gap now stands below 1%.
2. And VI and BSNL seem to further decrease in the market.
3. While the numbers in total telecom users have increased it is still very lesser and a large number of people fall in others category.

```python
import numpy as np
companies = ['Airtel', 'Jio', 'VI', 'BSNL', 'Others'] # Company names

# Set the width of the bars
bar_width = 0.35

r1 = np.arange(len(Bihar_start))
r2 = [x + bar_width for x in r1]

# Create the side-by-side bar graph
plt.figure(figsize=(10, 6))
plt.bar(r1, Bihar_start, color='aqua', width=bar_width, edgecolor='grey', label='Start of Year')
plt.bar(r2, Bihar_end, color='green', width=bar_width, edgecolor='grey', label='End of Year')

# Add xticks on the middle of the group bars
plt.xlabel('Network Providers', fontweight='bold')
plt.ylabel('Percentage of Subscribers', fontweight='bold')
plt.xticks([r + bar_width/2 for r in range(len(Bihar_start))], companies)

# Add a legend
plt.legend()

# Show the plot
plt.title('Comparison of % of Subscribers at Start and End of Year 2023 in Bihar(Jharkhand)')
plt.show()
```

The above graph compares the percentage of users at start of 2023 and by the end of 2023

1. Here also the same story, with Jio and Airtel increasing,
2. And VI and BSNL on a decline
3. And the number in non-subscribed users have fallen.
