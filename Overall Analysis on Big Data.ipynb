```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Airtel = pd.read_csv('Airtel21-23.csv', index_col='State')
Jio = pd.read_csv('Jio21-23.csv', index_col='State')
BSNL = pd.read_csv('BSNL21-23.csv', index_col='State')
VI = pd.read_csv('Vi21-23.csv', index_col='State')
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

## Let's do some magic on Airtel dataset

```python
Airtel.loc['Uttar Pradesh'] = Airtel.loc['U.P.(E)'] + Airtel.loc['U.P.(W)']
Airtel.loc['Maharashtra'] = Airtel.loc['Maharashtra'] + Airtel.loc['Mumbai']
Airtel.loc['West Bengal'] = Airtel.loc['West Bengal'] + Airtel.loc['Kolkata']
Airtel.loc['North East'] = Airtel.loc['North East'] + Airtel.loc['Assam']
```

```python
Airtel = Airtel.drop(['U.P.(W)','U.P.(E)', 'Mumbai', 'Kolkata', 'Assam', 'Total'])
```

```python
Total = []
for col in Airtel.columns:
    Total.append(Airtel[col].sum())
```

```python
Airtel.loc['Total'] = Total
```

## And On Jio Dataset Now

```python
Jio.loc['Uttar Pradesh'] = Jio.loc['U.P.(E)'] + Jio.loc['U.P.(W)']
Jio.loc['Maharashtra'] = Jio.loc['Maharashtra'] + Jio.loc['Mumbai']
Jio.loc['West Bengal'] = Jio.loc['West Bengal'] + Jio.loc['Kolkata']
Jio.loc['North East'] = Jio.loc['North East'] + Jio.loc['Assam']
```

```python
plt.figure(figsize=(10, 6))
Airtel.loc['Total'].plot(kind= 'bar')
Airtel.loc['Total'].plot(kind= 'line', marker = 'o', color='green')

plt.title('Airtel users All over India')
plt.xlabel('Month wise')
plt.ylabel('Total telecom users')
plt.xticks(rotation = 90)
plt.ticklabel_format(style='plain', axis='y')
plt.ylim(340000000)
```

```python
plt.figure(figsize=(10, 6))
plt.barh(Jio.index[:18], Jio['Dec23'][:18], label = 'Jio')
plt.barh(Airtel.index[:18], Airtel['Dec23'][:18], label = 'Airtel')
plt.barh(VI.index[:18], VI['Dec23'][:18], label = 'VI')

plt.title('Telecom users in India by end of 2023')
plt.ylabel('State of India')
plt.xlabel('Total telecom users')
plt.ticklabel_format(style='plain', axis='x')
plt.legend()
```
